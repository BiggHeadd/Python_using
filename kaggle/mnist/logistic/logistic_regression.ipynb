```python
# -*-coding:utf-8-*-
# Edited by bighead 19-2-11

import pandas as pd
import numpy as np
from time import time
from utils import get_train_data, get_hog_feature, get_test_data
from sklearn.linear_model import LogisticRegression
```

```python
########## Load the train dataset
print("loading training data")
load_start = time()
Y, X = get_train_data("data/train.csv")
X = get_hog_feature(X)
print("finish, cost: {} seconds".format(time() - load_start))
```

```python
########## build and fit the logistic regression model from sklearn
model = LogisticRegression()
model.fit(X, Y)
```

```python
########## Load the test dataset
print("loading test data")
load_start = time()
X_test = get_test_data("data/test.csv")
X_test = get_hog_feature(X_test)
print("finish, cost: {} seconds".format(time() - load_start))
```

```python
########## Predict the testset
print("\npredict.....")
predict_start = time()
predict = model.predict(X_test)
print("finish, cost: {} seconds".format(time() - predict_start))
```

```python
########## Submit
submit_data = pd.DataFrame(columns=["Label"])
submit_data["Label"] = predict
submit_data.insert(0, "ImageID", range(1, 1+len(predict)))
submit_data.to_csv('submit.csv', index=False)
submit_data
```
